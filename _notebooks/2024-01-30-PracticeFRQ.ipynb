4. POJO and chnages to a POJO - 0.8
POJO is a Plain Old Java Object, it is not restricted by any special restrictions besides Java itself
It can be changed using a setter method
For example:
public class Person {
    private String name;
    
    public Person() {
        // Default constructor
    }
    
    public String getName() {
        return name;
    }
    
    public void setName(String name) {
        this.name = name;
    }
}

// creating a POJO
Person a = new Person();
// Setting a value of a POJO
a.setName("Jack");
System.out.println(person.getName()); // This will print out Jack
